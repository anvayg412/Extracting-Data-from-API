```python
bases=['USD','GBP','AUD']
for base in bases:
    url="https://API.exchangerate.host/latest?base="+base
    response_API_rates_new=requests.get(url)
    data=response_API_rates_new.json()
    for currency,rates in data['rates'].items():
        print(currency,rates)
    
        
```

```python
print("Select from following list")
print("\n")
url="https://API.exchangerate.host/latest"
response_API_rates_new=requests.get(url)
data=response_API_rates_new.json()
lst=[]
for currency in data['rates']:
    lst.append(currency)
print(lst)
print("\n")

def currency_converter(base,convert):
    print("Base currency is: ",base)
    print("Convert to: ",convert)
    url="https://API.exchangerate.host/latest?base="+base
    response_API_rates_new=requests.get(url)
    data=response_API_rates_new.json()
    amount=float(input("Enter amount: "))
    rates=data['rates'][convert]
    result=amount*rates
    print(base,1,"=",convert,rates)
    print(base,amount,"to",convert,result)
        

base_currency=input("Enter base currency: ").upper()
convert_currency=input("Enter convert currency: ").upper()
currency_converter(base_currency,convert_currency)
    
```

```python
print("Select from following list")
print("\n")
url="https://API.exchangerate.host/latest"
response_API_rates_new=requests.get(url)
data=response_API_rates_new.json()
lst=[]
for currency in data['rates']:
    lst.append(currency)
print(lst)
print("\n")

def currency_converter(base,convert):
    print("Base currency is: ",base)
    print("Convert to: ",convert)
    url="https://API.exchangerate.host/latest?base="+base
    response_API_rates_new=requests.get(url)
    data=response_API_rates_new.json()
    amount=float(input("Enter amount: "))
    for currency,rate in data['rates'].items():
        if currency==convert:
            result=rate*amount
            print(base,1,"=",convert,rates)
            print(base,amount,"to",convert,result)
        

base_currency=input("Enter base currency: ").upper()
convert_currency=input("Enter convert currency: ").upper()
currency_converter(base_currency,convert_currency)
    
```
